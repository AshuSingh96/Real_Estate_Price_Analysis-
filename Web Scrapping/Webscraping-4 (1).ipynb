```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

import pandas as pd


from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
```

```python

driver = webdriver.Chrome()

driver.maximize_window()
```

```python
#https://www.squareyards.com/sale/property-for-sale-in-
url = 'https://www.squareyards.com/sale/property-for-sale-in-mumbai'
driver.get(url)
```

```python
property_type = []
address = []
price_range = []
total_area = []
all_features = []
listing_type= []
```

```python
amenities = []
project_info_text = [] 
price = []
```

```python

for i in range(1,101):
    driver.get(f"https://www.squareyards.com/sale/property-for-sale-in-mumbai?page={i}")
    time.sleep(np.random.randint(4, 5)) 
    
    elements = driver.find_elements(By.XPATH, "//div[contains(@class, 'dseprojectdata') and contains(@class, 'tileNo_')]")
    time.sleep(np.random.randint(4, 5))  

    print(len(elements))  # To know how many elements (listings) are found on the page
    inc = 1

    for j in elements:
        try:
            driver.execute_script("arguments[0].scrollIntoView(true);", j)
            time.sleep(np.random.uniform(1, 2))  # Small delay for stability
            j.click()
            driver.switch_to.window(driver.window_handles[1])

            # Ensure the page has loaded fully
            WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, "body")))

            # Retrieve price
            try:
                price_element = WebDriverWait(driver, 15).until(
                    EC.presence_of_element_located((By.XPATH, "/html/body/div[4]/div[2]/div[1]/div[2]/div[1]/strong"))
                )
                price.append(price_element.text)
            except TimeoutException:
                price.append(None)
                print("Price not found for this listing")

            # Retrieve project information
            try:
                project_info_element = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.ID, "projectInformation"))
                )
                project_info_text.append(project_info_element.text)
            except TimeoutException:
                project_info_text.append(None)

            time.sleep(np.random.randint(2, 3))  # Small delay to allow all content to load

            # Retrieve amenities
            try:
                amenities_elements = driver.find_elements(By.XPATH, "/html/body/div[6]/div/div/div[1]/div[3]/div[2]/div/table")
                amenities_text = ""

                for idx in range(1, len(amenities_elements) + 1):
                    amenity_xpath = f"/html/body/div[6]/div/div/div[1]/div[3]/div[2]/div/table[{idx}]"
                    try:
                        amenity_element = WebDriverWait(driver, 5).until(
                            EC.presence_of_element_located((By.XPATH, amenity_xpath))
                        )
                        amenities_text += amenity_element.text + ", "
                    except TimeoutException:
                        amenities_text += "None, "  # Placeholder if an amenity is missing

                amenities.append(amenities_text.rstrip(", "))
            except:
                amenities.append("No amenities found")

            inc += 1

            driver.close()
            driver.switch_to.window(driver.window_handles[0])

        except Exception as e:
            print(f"Error processing element: {e}")
            continue

```

```python
print(len(price), len(amenities), len(project_info_text) , sep="\n")
```
